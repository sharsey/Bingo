```python
import random
```

```python
class bingo:
  def __init__(self):
    self.code = ""
    self.count = 0
    self.card = {
        "B":[],
        "I":[],
        "N":[],
        "G":[],
        "O":[],
    }
    
  def get_code(self):
    return self.code

  def set_code(self, code):
    self.code = code

  def get_card(self):
    return self.card

  def set_card(self, card):
    self.card = card

  def get_count(self):
    return self.count

  def set_count(self, count):
    self.count = count

  def __str__(self):
    return self.code
  
  def generate_card(self, cont):
    min = 1
    max = 16
    self.count += cont
    self.code = self.build_code(self.count)
    for letter in self.card:
      self.card[letter] = self.build_numbers(min, max)
      min += 15
      max += 15
      if letter == "N":
        self.card[letter][2] = "TABLA Nº "+self.code+" 26/08/2021"

  def build_numbers(self, beginning, end):
    value = []
    limit = 0
    while limit<5:
      number = random.randrange(beginning, end)
      if (number in value)==False:
        limit += 1
        value.append(number)
    return value
  
  def build_code(self, count):
    final = ""
    if (count<10):
      final = "0000"+ str (count)
    elif (count>=10 and count<100):
      final = "000"+ str (count)
    elif (count>=100 and count<1000):
      final = "00"+ str (count)
    elif (count>=1000 and count<10000):
      final = "0"+ str (count)
    elif(count>=10000):
      final = str(count)
    return final

  def print_card(self):
    for letter in self.card:
      print(letter, end="\t")
      for number in self.card[letter]:
          print(number, end="\t")
      print("\n")
    print("\n")

  def printf(self):
    letter_b = list(self.card.get('B'))
    letter_i = list(self.card.get('I'))
    letter_n = list(self.card.get('N'))
    letter_g = list(self.card.get('G'))
    letter_o = list(self.card.get('O'))
    print("---------------------------------------")
    print ("{:<8} {:<8} {:<8} {:<8} {:<8}".format('B','I','N','G','O'))
    for i in range(5):
      print ("{:<8} {:<8} {:<8} {:<8} {:<8}".format(str (letter_b[i]),str (letter_i[i]),str (letter_n[i]), str (letter_g[i]), str (letter_o[i])))
    print("---------------------------------------")
    ## print(str (self.letter_b[i])+"  "+ str (self.letter_i[i])+"  "+ str (self.letter_n[i])+"   "+str (self.letter_g[i])+"  "+ str(self.letter_o[i]))


      
    
```

```python
number = input("Ingrese las tablas de bingo a generar: ")
```

```python
lst = []
for i in range(int (number)):
  bingo1 = bingo()
  bingo1.generate_card(i+1)
  lst.append(bingo1.get_card())
  bingo1.printf()
```

```python
def html2pdf(html_path, pdf_path):
    """
    Convert html to pdf using pdfkit which is a wrapper of wkhtmltopdf
    """
    options = {
        'page-size': 'Letter',
        'margin-top': '0.35in',
        'margin-right': '0.75in',
        'margin-bottom': '0.75in',
        'margin-left': '0.75in',
        'encoding': "UTF-8",
        'no-outline': None,
        'enable-local-file-access': None
    }
    with open(html_path) as f:
        pdfkit.from_file(f, pdf_path, options=options)
```
